##### Be careful when copying arrays!!!

```python
a = np.array([1, 2, 3])
b = a
c = a.copy()
b[0] = 100
c[0] = 200
print(b)
print(a)
print(c)
print(a)
```

##### Reorganizing arrays

```python
before = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(before)

after = before.reshape((4, 2))
print(after)
```

```python
#Vertical stacking
v1 = np.array([1, 2, 3, 4])
v2 = np.array([5, 6, 7, 8])

np.vstack([v1, v2, v1, v2])
```

```python
#Horizontal stacking
hp1 = np.ones((2, 4))
hp2 = np.zeros((2, 2))

np.hstack((hp1, hp2))
```
